### Шаг 4

Пожалуйста, реализуйте функцию ```plot_sns_pclass(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе класса пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_pclass(data=None):
    c_list=data["class"].value_counts().index.tolist()
    for info, n in zip(c_list,range(1,len(c_list)+1)):
        category=data[data["class"] == info]
        f=sns.set(style="white")
        f=sns.set_context("talk")
        f=sns.relplot(x="fare",y="age",hue="alive",height=4,aspect=2,hue_order=["yes","no"],data=category)
        plt.title(f"Passenger class: {n}")
        plt.gcf().set_size_inches(8, 4)
        leg=f._legend
        leg.set_bbox_to_anchor([1.05,0.525])
    #Конец вашего кода
plot_sns_pclass(data=titanic)
````

### Шаг 5


Пожалуйста, реализуйте функцию ```plot_sns_sex(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе пола пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_sex(data=None):
    #Начало вашего кода
    c_list=data["sex"].value_counts().index.tolist()
    for info in reversed(c_list):
        category=data[data["sex"]==info]
        sns.set_context("talk")
        last=sns.relplot(x='fare',y="age",hue="alive",height=4,aspect=2,hue_order=["yes","no"],data=category)
        plt.title(f"Passenger sex: {info}")
        plt.gcf().set_size_inches(8,4)
        leg=last._legend
        leg.set_bbox_to_anchor([1.06,0.525])
    #Конец вашего кода

plot_sns_sex(data=titanic)
````

### Шаг 6


Пожалуйста, реализуйте функцию ```plot_sns_town(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе города пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_town(data=None):
    #Начало вашего кода
    c_list=data["embark_town"].value_counts().index.tolist()
    c_list.sort()
    for info in c_list:
        category=data[data["embark_town"]==info]
        sns.set_context("talk")
        last=sns.relplot(x="fare",y="age",hue='alive',height=3,hue_order=["yes","no"],legend=True,aspect=3,data=category)
        plt.title(f"Passenger town: {info}")
        plt.gcf().set_size_inches(8,4)
        leg=last._legend
        leg.set_bbox_to_anchor([1.07,0.5])
    #Конец вашего кода

plot_sns_town(data=titanic)
````
