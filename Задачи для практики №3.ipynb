4. Пожалуйста, произведите следующий график, показанный ниже (ширина = 0,5):



````python
# Начало вашего кода
def marriage(data):
    yes=data["married"].value_counts()[0]
    no=data["married"].value_counts()[1]
    plt.bar(x=["yes","no"],height=[yes,no],width=0.5);
    plt.title("Counts marriage status");
    plt.ylabel("Counts");
    
marriage(pd_data)
# Конец вашего кода
````

5. Пожалуйста, произведите следующий график, показанный ниже (ширина = 0,5):



````python
# Начало вашего кода
def experience(data):
    senior=data["experience"].value_counts()[0]
    middle=data["experience"].value_counts()[1]
    fresh=data["experience"].value_counts()[2]
    plt.bar(x=[">5","3-5","0-2"],height=[senior,middle,fresh],width=0.5);
    plt.title("Counts experience years");
    plt.ylabel("Counts");
    
experience(pd_data)
# Конец вашего кода
````

8. Пожалуйста, произведите следующий график, показанный ниже (размер figure установлен на (10,7)):




````python
# Начало вашего кода
male=[pd_data[pd_data["gender"]=="male"],"Male"]
female=[pd_data[pd_data["gender"]=="female"],"Female"]
married=[pd_data[pd_data["married"]=="yes"],"Married"]
single=[pd_data[pd_data["married"]=="no"],"Single"]

info_list=[male,female,married,single]

exp=["0-2","3-5",">5"]

fig=plt.figure(figsize=(10,7))
axs=[]
axs.append(fig.add_subplot(2,2,1))
axs.append(fig.add_subplot(2,2,2))
axs.append(fig.add_subplot(2,2,3))
axs.append(fig.add_subplot(2,2,4))

for ax,cat in zip(axs,info_list):
    col=[]
    for i in exp:
        col.append(cat[0][cat[0]["experience"]==i]["salary"])
    ax.boxplot(col)
    ax.set_title(cat[1])
    ax.set_yticks(range(10000,50001,5000))
    ax.set_xlabel("Experience level")  
    ax.set_xticklabels(labels=exp ,minor=False,rotation=45)
    plt.tight_layout()  

# Конец вашего кода
````

9. Пожалуйста, произведите следующий график, показанный ниже (размер figure установлен на (10,7)):



````python
# Начало вашего кода
fig2=plt.figure(figsize=(10,7))
ax_bar=[]
ax_bar.append(fig2.add_subplot(2,2,1))
ax_bar.append(fig2.add_subplot(2,2,2))
ax_bar.append(fig2.add_subplot(2,2,3))
ax_bar.append(fig2.add_subplot(2,2,4))
for ax,cat in zip(ax_bar,info_list):
    for i,num in zip(exp,range(len(exp))):
        color=["r","b","g"]
        ax.bar(x=i,height=len(cat[0][cat[0]["experience"]==i]["salary"]),color=color[num],edgecolor="k")
        ax.set_title(cat[1]);
        ax.set_xlabel("Experience level");
        ax.set_xticklabels(labels=exp ,rotation=45);
        plt.tight_layout();  
               
# Конец вашего кода
````
