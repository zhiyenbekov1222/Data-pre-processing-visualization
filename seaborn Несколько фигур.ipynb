### Шаг 3

Сделайте фигуру, как показано ниже. Пожалуйста, будьте осторожны с тем, какие столбцы из DataFrame изображены на рисунке.

### Ожидаемый результат:



````python
#Начало вашего кода
t_list=titanic.columns.tolist()
int_list=["sex"]
for i in t_list:
    if titanic[i].dtypes=="int64" or titanic[i].dtypes=="float64" :
        int_list.append(i)
        
sns.pairplot(titanic[int_list],diag_kind="kde",hue="sex",corner=True);
#Конец вашего кода
````

### Шаг 4

Сделайте фигуру, как показано ниже (высота фигуры = 5 и ширина фигуры = 5)

### Ожидаемый результат:



````python
#Начало вашего кода
f=sns.set(style="white")
f=sns.set_context("talk")
f=sns.FacetGrid(titanic,col="alone",row="sex",hue="alive",size=5,hue_order=["yes","no"])
f.map_dataframe(sns.scatterplot,"age","fare").add_legend()
leg=f._legend
leg.set_bbox_to_anchor([1.09,0.48])
plt.tight_layout()
#Конец вашего кода
````

### Шаг 5

Сделайте фигуру, как показано ниже (высота фигуры = 5 и ширина фигуры = 5)

### Ожидаемый результат:



````python
#Начало вашего кода
s5=sns.set(style="darkgrid")
s5=sns.set_context("talk")
s5=sns.FacetGrid(titanic,row="alone",col="pclass",row_order=[True,False],height=5,hue="alive",hue_order=["yes","no"]);
s5.map_dataframe(sns.swarmplot,"sex","age").add_legend();
#Конец вашего кода
````
