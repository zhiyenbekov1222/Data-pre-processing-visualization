Пожалуйста, реализуйте функцию ```statistics_by_group(data,column)```, которая возвращает статистику по группе ```column``` для столбца ```"SalePrice"``` в ```DataFrame```. Статистика содержит среднее, стандартное отклонение и медианные значения по группе. Статистика сортируется по среднему значению, медиане и стандартному отклонению, по убыванию, убыванию и возрастанию соответственно.

````python
#Начало вашего кода
def statistics_by_group(data,column="SalePrice"):
    
    data_need1=data.groupby(column)["SalePrice"].mean()
    pd_data1=pd.DataFrame(data_need1)
    pd_data1.columns=["mean"]
    
    data_need2=data.groupby(column)["SalePrice"].std()
    pd_data2=pd.DataFrame(data_need2)
    pd_data2.columns=["std"]
    
    data_need3=data.groupby(column)["SalePrice"].median()
    pd_data3=pd.DataFrame(data_need3)
    pd_data3.columns=["median"]
    
    pd_result=pd.concat([pd_data1,pd_data2,pd_data3],axis=1)
    pd_result=pd_result.sort_values(["mean","std","median"],ascending=[False,True,False])
    
    return pd_result
#Конец вашего кода
````

````python
result=statistics_by_group(data=pd_data,column="OverallQual")
print(result)
````

````python
#Начало вашего кода
def extract_important(data, matrix, threshold = 0.5):
    list_index=[]
    for i in range(len(matrix)):
        if matrix["SalePrice"][i]>threshold:
            list_index.append(matrix.index[i])
    return data[list_index]
#Конец вашего кода
````
