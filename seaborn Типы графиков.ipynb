### Шаг 5

Пожалуйста, реализуйте функцию ```plot_sns_age_pclass(data=None)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе класса пассажира.

**Замечания: высота фигуры = 4 и ширина фигуры = 8**

````python
def plot_sns_age_pclass(data=None):
    #Начало вашего кода
    d=data["class"].value_counts().index.tolist()
    reserve=d[0]
    d[0],d[1],d[2]=d[1],d[2],reserve
    for info,n in zip(d,range(1,len(d)+1)):
        cl_data=data[data["class"]==info]
        fig=sns.catplot(x="sex",y="age",hue="alive",hue_order=["yes","no"],height=3,aspect=3,data=cl_data)
        plt.title(f"Passenger class: {n}");
        plt.gcf().set_size_inches(8,4);
        leg=fig._legend
        leg.set_bbox_to_anchor([1.08,0.5])
    #Конец вашего кода

plot_sns_age_pclass(data=titanic)
````

### Шаг 7

Пожалуйста, реализуйте функцию plot_counts(data=None), которая отображает несколько фигур, каждая фигура содержит визуализацию на основе ```param```.

````python
def plot_counts(data=None,param=[],fontsize=15):
    #Начало вашего кода
    i_list=[]
    fig,ax=plt.subplots(len(param),1)
    for item,i in zip(param,range(len(param))):
        items=sns.set_theme(style="whitegrid")
        items=sns.countplot(x="alive",hue=item,data=data,ax=ax[i])
        items.set_title(f"Counts based on {item}",fontsize=fontsize)
        plt.gcf().set_size_inches(6.2,7)
        plt.tight_layout()
    #Конец вашего кода

plot_counts(data=titanic,param=["pclass","sex"])
````

### Шаг 8

Пожалуйста, реализуйте функцию ```plot_dists(data=None,param=[],fontsize=15)```, которая отображает несколько фигур, каждая фигура содержит визуализацию на основе ```param```.


````python
def plot_dists(data=None,param=[],fontsize=15):
    #Начало вашего кода
    fig,ax=plt.subplots(len(param),1);
    for item,i in zip(param, range(len(param))):
        items=sns.distplot(data[item],ax=ax[i]);
        items.set_title(item + " distribution",fontsize=fontsize)
        items.set_ylabel(None)
        plt.gcf().set_size_inches(5.3,8)
        plt.tight_layout()
    #Конец вашего кода

plot_dists(data=titanic,param=["fare","age"])
````
