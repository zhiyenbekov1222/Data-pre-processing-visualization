### Шаг 5

Пожалуйста, создайте ```pie``` с помощью ```rand_nums``` (созданную ранее).

**Подсказка: проверьте входной аргумент ```explode```.** 
Должен быть установлен на ```0.1```.

````python
#Начало вашего кода
list_pie=[]
for i in rand_nums:
    if i not in list_pie:
        list_pie.append(i)
        
count=0
p_list=[]
c_dict={}
for u in list_pie:
    for n in rand_nums:
        if u==n:
            count +=1
    p_list.append(count)
    count=0
distance=0.1
separate = (distance, distance, distance, distance, distance, distance, distance, distance, distance, distance)
plt.pie(x=p_list,labels=list_pie,autopct=lambda x:f"{x:.1f}%",pctdistance=0.8,
        rotatelabels=False,explode=separate);
plt.title("Pie chart");
#colors=["b","g","r","m","w","m","r","y","c","k"]
#Конец вашего кода
````
